

# Bar Graph
### Average Calories Burned During Workouts

This bar graph shows the average calories burned for various workout types, . Each bar represents a specific workout type along with the corresponding average calories burned per session.

#### Purpose
The purpose of this graph is to compare the calorie-burning potential of each workout type.




```python
"""The Average Burned Calories were calculated for Workout type which is grouped  groupby and coloumns were renamed as calories_burned_avg"""

total_work_type = gym_df.groupby('Workout_Type')['Calories_Burned'].mean().reset_index(name='Calories_Burned_avg')
#Defining Column for new grouped data frame

total_work_type.columns = ['Workout_Type', 'Calories_Burned_avg']

#Showing The dataframe of average calories burned 

total_work_type.head()

```

```python
#Ploting a  bar chart for average calories burnt to their respective workout types

def bar_graph(total_work_type):
    # adjusting the width of the plot figure
   
    plt.figure(figsize=(25, 8))
    #ploting bar chart from averaged value

    plt.bar(x=total_work_type['Workout_Type'],height=total_work_type['Calories_Burned_avg'],label=total_work_type['Workout_Type'],color=['red','darkgreen','yellow','pink'],width=[0.4])
    
    #Giving Annoted Value for bar top label
    
    for index, value in enumerate(total_work_type['Calories_Burned_avg']):
        plt.text(index, value + 0.5, str(round(value,1)), ha='center', va='bottom')
        #Giving Title for Bar Chart
    plt.title("Distribution of Workout Types")
    #defining the label legend 
    
    plt.legend()

    #labeling x axis
    
    plt.xlabel('Work out types')

    #labeling Y-axis
    
    plt.ylabel('Average Calories Burnt')

    #displays the output
    plt.show()

#calling The graph Function
bar_graph(total_work_type)
```

# Line Graph
#### Line graphs shows the continous data and it allows you to see increase,decrease and fluctuation of data
## Purpose
##### To track where the trends and pattern need to highlighted and categorizing the value for same dataset


```python
def line_graph(male_data,female_data):
        """ 
            -The Grouped Data Frame of Gym dataset wer  grouped by Gender ,Workout type and workout duration as  named counts 
            -Splitting the workout data for two genders ,which  is male and female 
            -Plotting the value for both genders using line plot.
            -The lingraph uses two arguments  which is male data and  female data, which called below
        """
        #plotting for male data 
        plt.plot(male_data['Workout_Type'], male_data['Counts'],marker='o',label='Male')

        #plotting for female data
        plt.plot(female_data['Workout_Type'], female_data['Counts'],marker='o',label='Female')

        #plotting the x-axis
        plt.xlabel('Workout Types')

        #plotting the y-axis
        plt.ylabel('Time Duration in Hour ')

        #giving title for line graph
        plt.title('People average workout durations and their  Workout Types')

        #styling with grid
        plt.grid(True)

        #showing the labels
        plt.legend()

        #displaying the output
        plt.show()    

        return
#Creating a Gender Data Frame 
gender_df=pd.DataFrame(gender_group)

#Male grouped  DataFrame using Condition
male_df=gender_df[gender_df['Gender']=='Male']

#female Grouped DataFrame Using Condition
female_df=gender_df[gender_df['Gender']=='Female']

#calling the line graph function
line_graph(male_df,female_df)
```
